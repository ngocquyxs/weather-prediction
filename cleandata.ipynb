```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/IOT/18/iot-wp-default-rtdb-export (1).csv')
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.reset_index(inplace=True)
```

```python
df
```

```python
df['TEMP'] = pd.to_numeric(df['TEMP'], errors='coerce')
```

```python
df['PRCP'] = pd.to_numeric(df['PRCP'], errors='coerce')

df['PRCP'] = df['PRCP'] / 100

print(df)
```

```python
df['TMAX'] = float('nan')
df['TMIN'] = float('nan')
```

```python
df
```

```python
df.at[0, 'TMAX'] = df.at[0, 'TEMP']
df.at[0, 'TMIN'] = df.at[0, 'TEMP']
```

```python
df
```

```python
for i in range(1, len(df)):
    condition = df.iloc[i, df.columns.get_loc('TEMP')] >= df.iloc[i - 1, df.columns.get_loc('TMAX')]
    df.iloc[i, df.columns.get_loc('TMAX')] = df.iloc[i, df.columns.get_loc('TEMP')] if condition else df.iloc[i - 1, df.columns.get_loc('TMAX')]
```

```python
df
```

```python
for i in range(1, len(df)):
    condition = df.iloc[i, df.columns.get_loc('TEMP')] <= df.iloc[i - 1, df.columns.get_loc('TMIN')]
    df.iloc[i, df.columns.get_loc('TMIN')] = df.iloc[i, df.columns.get_loc('TEMP')] if condition else df.iloc[i - 1, df.columns.get_loc('TMIN')]
```

```python
df
```

```python
df = df.drop('TEMP', axis=1)
```

```python
df.to_csv('output_file.csv', index=False)
```
